#Tutorial 4 - Automate Analysis Functions

```python
def retention_rate(dataframe, column_names):

  retained = dataframe[dataframe['converted'] == True]\
             [dataframe['is_retained'] == True]\
             .groupby(column_names)['user_id'].nunique()

  converted = dataframe[dataframe['converted'] == True]\
             .groupby(column_names)['user_id'].nunique()

  retention_rate = retained / converted

  retention_rate = retention_rate.fillna(0)

  return retention_rate
```

```python
daily_retention = retention_rate (marketing, ['date_subscribed','subscribing_channel'])

daily_retention = pd.DataFrame(daily_retention.unstack(level=1))

daily_retention.head()
```

#Tutorial 6 - Resolving Inconsistency

```python
# ensure all date are in standard date format
marketing['date_served']=pd.to_datetime(marketing['date_served'])
print(marketing['date_served'].head())

# Isolate the rows where marketing channel is House Ads
house_ads = marketing[marketing['marketing_channel'] == 'House Ads']
print(house_ads.head())
```

```python
# defining conversion rate function as in tutorial 4
def conversion_rate(dataframe, column_names):

  column_total = dataframe.groupby(column_names)['user_id'].nunique()

  column_cov = dataframe[dataframe['converted'] == True]\
             .groupby(column_names)['user_id'].nunique()

  conversion_rate = column_cov/column_total

  conversion_rate = conversion_rate.fillna(0)

  return conversion_rate
```

```python
# Calculate pre-error conversion rate
house_ads_pre = house_ads[house_ads['date_served'] < '2018-01-11']
lang_conv_pre = conversion_rate(house_ads_pre, ['language_displayed'])
print(lang_conv_pre)
```

```python
# Index other language conversion rate against English
spanish_index_pre = lang_conv_pre['Spanish']/lang_conv_pre['English']
arabic_index_pre = lang_conv_pre['Arabic']/lang_conv_pre['English']
german_index_pre = lang_conv_pre['German']/lang_conv_pre['English']

print("Spanish index pre-bug:", spanish_index_pre)
print("Arabic index pre-bug:", arabic_index_pre)
print("German index pre-bug:", german_index_pre)
```

```python
# Group house_ads by date and language
converted = house_ads.groupby(['date_served',
                          'language_preferred'])\
                        .agg({'user_id':'nunique',
                              'converted':'sum'})
display(converted)
```

```python
# Unstack dataframe
converted_df = pd.DataFrame(converted.unstack(level=1))
display(converted_df) # notice we use converted_df instead of converted
```

```python
# Create English converted customers column for post-error (affected) period
converted_df['english_conv_post'] = converted_df.loc['2018-01-11':'2018-01-31'][('converted','English')]
display(converted_df)
```

```python
# Create expected converted customers for each language for affected period if there is no bug
converted_df['expected_spanish_post'] = converted_df['english_conv_post']*spanish_index_pre
converted_df['expected_arabic_post'] = converted_df['english_conv_post']*arabic_index_pre
converted_df['expected_german_post'] = converted_df['english_conv_post']*german_index_pre
display(converted_df)
```

```python
# Multiply number of users by the expected conversion rate
converted_df['expected_spanish_conv_post'] = converted_df['expected_spanish_post']/100*converted_df[('user_id','Spanish')]
converted_df['expected_arabic_conv_post'] = converted_df['expected_arabic_post']/100*converted_df[('user_id','Arabic')]
converted_df['expected_german_conv_post'] = converted_df['expected_german_post']/100*converted_df[('user_id','German')]
display(converted_df)
```

```python
# Use .loc to slice only the relevant dates
bug_period = converted_df.loc['2018-01-11':'2018-01-31']

# Sum expected subscribers for each language
expected_subs = bug_period['expected_spanish_conv_post'].sum() + \
             bug_period['expected_arabic_conv_post'].sum() + bug_period['expected_german_conv_post'].sum()

# Calculate how many subscribers we actually got
actual_subs = bug_period[('converted','Spanish')].sum() + bug_period[('converted','Arabic')].sum() + \
             bug_period[('converted','German')].sum()

# Subtract how many subscribers we got despite the bug
lost_subs = expected_subs - actual_subs
print(lost_subs)
```

#Tutorial 9 - A/B Test & Segmentation

```python
# Creata life function to calculate life for group a and b
import numpy as np

def lift(a,b):
    # Calcuate the mean of a and b
    a_mean = np.mean(a)
    b_mean = np.mean(b)

    # Calculate the lift using a_mean and b_mean
    lift = (b_mean-a_mean)/a_mean

    return str(round(lift*100, 2)) + '%'
```

```python
# Perform a t significance test for group a and b
from scipy.stats import ttest_ind

def t_statistic(a, b):
  a = a.astype(float)
  b = b.astype(float)

  t = ttest_ind(a, b)

  return t
```

```python
for language in np.unique(marketing['language_displayed'].values):

  print(language)

  language_data = marketing[(marketing['marketing_channel'] == 'Email')&
                            (marketing['language_displayed'] == language)]

  subscribers = language_data.groupby(['user_id', 'variant'])['converted'].max()

  subscribers = pd.DataFrame(subscribers.unstack(level=1))

  control = subscribers['control'].dropna()

  personalization = subscribers['personalization'].dropna()

  print(lift(control, personalization))
  print(t_statistic(control, personalization))
```

```python
def ab_segmentation(segment):
  # Build a for loop for each segment in marketing
  for subsegment in np.unique(marketing[segment].values):
      print(subsegment)

      # Limit marketing to email and subsegment
      df = marketing[marketing[segment] == subsegment]

      subscribers = df.groupby(['user_id', 'variant'])['converted'].max()
      subscribers = pd.DataFrame(subscribers.unstack(level=1))
      control = subscribers['control'].dropna()
      personalization = subscribers['personalization'].dropna()

      print('lift:', lift(control, personalization))
      print('t-statistic:', t_statistic(control, personalization), '\n\n')
```
